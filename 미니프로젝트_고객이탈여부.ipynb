### 7. df3의 모든 컬럼에 대해 결측치를 확인하는 코드를 작성하고 결측치를 처리하세요.
* df3을 df4에 저장하세요.
* df4의 결측치가 40% 이상인 컬럼은 컬럼을 삭제하세요.
* df4의 결측치가 40% 미만인 컬럼은 결측치가 있는 row를 삭제하세요.

```python
# 여기에 답안코드를 작성하세요.

# df3의 모든 컬럼에 대해 결측치 확인
missing_counts = df3.isnull().sum()
print("결측치 개수:")
print(missing_counts)

# df3을 df4에 저장
df4 = df3.copy()

# 결측치 비율 계산
missing_percentage = df4.isnull().mean()

# 40% 이상 결측치가 있는 컬럼 삭제
cols_to_drop = missing_percentage[missing_percentage >= 0.4].index
df4.drop(columns=cols_to_drop, inplace=True)

# 40% 미만 결측치가 있는 컬럼의 결측치가 있는 행 삭제
df4.dropna(axis=0, subset=df4.columns[df4.isnull().mean() < 0.4], inplace=True)

# 결과 확인 (선택 사항)
print("결측치 처리 후 df4의 shape:", df4.shape)
print("결측치 개수 (처리 후):")
print(df4.isnull().sum())



```

```python
# 여기에 답안코드를 작성하세요.


import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# 1. tenure에 대한 히스토그램 시각화
plt.figure(figsize=(10, 6))
sns.histplot(df4['tenure'], bins=30, kde=True)
plt.title('Distribution of Tenure (Service Duration)')
plt.xlabel('Tenure (Months)')
plt.ylabel('Frequency')
plt.show()

# 2. tenure를 x 값으로 churn을 hue 값으로 사용하여 kdeplot 시각화
plt.figure(figsize=(10, 6))
sns.kdeplot(data=df4, x='tenure', hue='Churn', common_norm=False, fill=True)
plt.title('KDE Plot of Tenure by Churn')
plt.xlabel('Tenure (Months)')
plt.ylabel('Density')
plt.show()

# '서비스 사용기간이 길어질 수록 이탈이 적다'에 대해 'O'인지 'X'인지 출력
mean_tenure_churned = df4[df4['Churn'] == 1]['tenure'].mean()
mean_tenure_not_churned = df4[df4['Churn'] == 0]['tenure'].mean()

if mean_tenure_churned < mean_tenure_not_churned:
    print("서비스 사용기간이 길어질 수록 이탈이 적다: O")
else:
    print("서비스 사용기간이 길어질 수록 이탈이 적다: X")

# 3. 'tenure', 'MonthlyCharges', 'TotalCharges' 간의 상관관계를 확인하여 heatmap으로 시각화
correlation_matrix = df4[['tenure', 'MonthlyCharges', 'TotalCharges']].corr()

plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Heatmap')
plt.show()

# 가장 높은 상관계수 값 출력
highest_corr = correlation_matrix.unstack().sort_values(ascending=False)
# 자기 상관계수는 제외 (1.0)
highest_corr = highest_corr[highest_corr < 1.0]
print("가장 높은 상관계수:", highest_corr.idxmax(), ":", highest_corr.max())

```

### 11. df5에 대해 Scikit-learn의 train_test_split 함수로 훈련, 검증 데이터를 분리하세요.
* 입력 : X, y (y에는 churn을 저장하고 X에는 churn을 제외한 나머지를 저장하세요)
* Train : Test 비율 = 8:2
* y Class 비율에 맞게 나누는 옵션을 추가하세요.
* random_state=42 로 설정하세요.
* 결과 : X_train, X_valid, y_train, y_valid에 저장하세요.

```python
df5.dtypes
```

```python
# 여기에 답안코드를 작성하세요.
from sklearn.model_selection import train_test_split

# df5에서 X와 y 정의
X = df5.drop(columns=['Churn'])  # 'Churn' 컬럼을 제외한 나머지
y = df5['Churn']  # 'Churn' 컬럼

# train_test_split으로 훈련 데이터와 검증 데이터 분리
X_train, X_valid, y_train, y_valid = train_test_split(
    X, y,
    test_size=0.2,  # 20%를 검증 데이터로 사용
    stratify=y,     # y 클래스 비율에 맞게 나누기
    random_state=42  # 랜덤 시드 설정
)

# 결과 확인 (선택 사항)
print("훈련 데이터 X의 shape:", X_train.shape)
print("검증 데이터 X의 shape:", X_valid.shape)
print("훈련 데이터 y의 shape:", y_train.shape)
print("검증 데이터 y의 shape:", y_valid.shape)
```

### 12. MinMaxScaler 함수를 'scaler'로 정의하고 데이터를 정규화하세요.

```python
# 여기에 답안코드를 작성하세요.

from sklearn.preprocessing import MinMaxScaler

# MinMaxScaler 객체 생성
scaler = MinMaxScaler()

# X_train 데이터를 정규화
X_train_scaled = scaler.fit_transform(X_train)

# X_valid 데이터를 정규화 (훈련 데이터의 스케일을 사용)
X_valid_scaled = scaler.transform(X_valid)

# 정규화된 데이터를 DataFrame으로 변환
X_train_scaled = pd.DataFrame(X_train_scaled, columns=X_train.columns)
X_valid_scaled = pd.DataFrame(X_valid_scaled, columns=X_valid.columns)

# 결과 확인 (선택 사항)
print("정규화된 훈련 데이터 X의 shape:", X_train_scaled.shape)
print("정규화된 검증 데이터 X의 shape:", X_valid_scaled.shape)

```

### 15. 고객 이탈여부를 예측하는 딥러닝 모델을 만들려고 합니다.
### 아래 가이드에 따라 모델링하고 학습을 진행하세요.
* Tensoflow framework를 사용하여 딥러닝 모델을 만드세요.
* 히든레이어(hidden layer) 2개이상으로 모델을 구성하세요.
* dropout 비율 0.2로 Dropout 레이어 1개를 추가해 주세요.
* 하이퍼파라미터 epochs: 30, batch_size: 16으로 설정해주세요.
* 각 에포크마다 loss와 metrics 평가하기 위한 데이터로 X_valid, y_valid 사용하세요.
* 학습정보는 history 변수에 저장해주세요

```python
# 여기에 답안코드를 작성하세요.

# 딥러닝 모델 정의
model = Sequential()

# 입력 레이어
model.add(Dense(64, activation='relu', input_shape=(X_train_scaled.shape[1],)))
model.add(Dropout(0.2))  # Dropout 레이어 추가

# 히든 레이어 1
model.add(Dense(32, activation='relu'))
model.add(Dropout(0.2))  # Dropout 레이어 추가

# 히든 레이어 2
model.add(Dense(16, activation='relu'))

# 출력 레이어 (이탈 여부는 이진 분류)
model.add(Dense(1, activation='sigmoid'))

# 모델 컴파일
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

# 조기 종료 콜백 정의 (선택 사항)
early_stopping = EarlyStopping(monitor='val_loss', patience=3, restore_best_weights=True)

# 모델 학습
history = model.fit(
    X_train_scaled,
    y_train,
    validation_data=(X_valid_scaled, y_valid),
    epochs=30,
    batch_size=16,
    callbacks=[early_stopping]
)

# 결과 확인 (선택 사항)
print("모델 학습 완료.")
```
